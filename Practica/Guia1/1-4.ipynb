## Ejercicio 4
Genere y grafique una señal senoidal discreta con frecuencia 5 Hz y duración 1 seg. Para ello utilice las siguientes frecuencias de muestreo: 100, 25, 10, 4, 1 y 0,5 Hz. Analice el resultado. ¿En qué casos la canitdad de ciclos que observa se corresponde con una sinusoidal de 5 Hz? ¿A qué se deben las discrepancias encontradas?

```python
import numpy as np
import matplotlib.pyplot as plt

def senoidal(tini, tfin, fm, fs, fase):
    T = 1.0 / fm
    t = np.arange(tini,tfin,T)
    return t, np.sin(2 * np.pi * fs * t + fase)

# Generar muestreos
s = [0]*6
s[0] = senoidal(0, 1, 100, 5, 0)    # fm = 100 Hz
s[1] = senoidal(0, 1, 25 , 5, 0)    # fm = 25 Hz
s[2] = senoidal(0, 1, 10 , 5, 0)    # fm = 10 Hz
s[3] = senoidal(0, 1, 4  , 5, 0)    # fm = 4 Hz
s[4] = senoidal(0, 1, 1  , 5, 0)    # fm = 1 Hz
s[5] = senoidal(0, 1, 0.5, 5, 0)    # fm = 0.5 Hz

# Graficar
fig,ax = plt.subplots(6,sharex=True)
for i in range(6):
    ax[i].grid()
    ax[i].plot(s[i][0],s[i][1])

plt.show()
```

### Respuesta

Los unicos casos en los que se observa una cantidad de ciclos correspondiente a una señal de 5Hz es para las frecuencias de muestreo de 100Hz y 25Hz. En el caso con frecuencia de muestreo 10Hz nos encontramos en el limite de la condicion dada por el _teorema de muestreo_ ($f_m \geq 2f_s$), por lo que no se alcanzan a apreciar los ciclos. En los otros casos (4Hz, 1Hz, 0.5Hz) no llegamos cumplir la condicion del teorema de muestreo, por lo que es imposible distinguir la señal muestreada o ni siquiera podemos graficar una señal
