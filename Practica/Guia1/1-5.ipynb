## Ejercicio 5
Genere y grafique una señal senoidal con frecuencia 4000 Hz y duración 2 seg., utilizando una frecuencia de meustre 129 Hz. Grafique el resultado y estime la frecuencia de la onda sinusoidal que se observa en la figura. Analice y obtenga conclusiones.

```python
import numpy as np
import matplotlib.pyplot as plt

# generar senial
tini = 0
tfin = 2
fs = 4000
fm = 129

# muestrear
t = np.linspace(tini,tfin, fm*2, endpoint=False)
s = np.sin(2*np.pi*fs*t)

t_prueba = np.linspace(tini,tfin, 10000*2, endpoint=False)
s_prueba = np.sin(2*np.pi*fs*t_prueba)

# graficar
fig, ax = plt.subplots(2,sharex=True)
fig.set_figwidth(20)
ax[0].grid()
ax[1].grid()
ax[0].plot(t,s,color='red')
ax[1].plot(t_prueba,s_prueba, color='green')
plt.show()
```

### Respuesta
La señal grafica tiene una frecuencia de 1 Hz, mientras que la señal que generamos y luego muestreamos tiene una frecuencia de 4000 Hz. La confusión se debe al fenómeno de _aliasing_.
