## Ejercicio 3
Determine los valores numericos (y unidades correspondientes) de amplitud (A), fase (phi), frecuencia (fs) y periodo de muestreo (Tm), que fueron utilizados para generar la señal mediante el siguiente codigo
```
t = 0:Tm:0.1-Tm;
x(t) = A*sin(2*pi*fs + phi);
plot(t,x);
```
Para determinar el valor de la fase tenga  en cuenta la relacion $\phi = -2 \pi f_s t_1$, donde $t_1$ indica el retardo temporal en segundos (en este caso el retardo temporal se peude determinar encontrando el primer cruce por cero de la sinusoidal).

### Solucion

Amplitud: $A = 3$ (valores absolutos maximos) \
periodo de muestreo: $T_m = \frac{0.01}{8} = 0.00125s$ (muestras entre 0.09 y 0.1) \
frecuencia muestral: $f_m = \frac{1}{T_m} = \frac{1}{0.00125} = 800Hz$ \
frecuencia de la señal: $f_s = \frac{2}{0.1} = 20Hz$ (2 ciclos en 0.1) \
fase: \
$\phi = -2 \pi f_s t_1$ \
$\phi = -2 \pi 20 (5*0.00125)$ (5 muestras hasta el primer cruce por cero de la sinusoidal) \
$\phi = -0.25 \pi$

````python
import numpy as np
import matplotlib.pyplot as plt

A = 3           # amplitud
Tm = 0.00125    # periodo muestral
fs = 20        # frecuencia de la senial
phi = -0.25*np.pi # fase

t = np.arange(0, 0.1, Tm)
s = A*np.sin(2*np.pi*fs*t + phi)

fig, ax = plt.subplots(1)
fig.set_figwidth(20)

ax.stem(t,s)
plt.show()
````
