## Ejercicio 1

Para cada uno de los siguientes sistemas determine si son causales, lineales, invariantes en el tiempo y si poseen memoria. En cada caso grafique la salida del sistema $y[n]$ para una entrada dada.

1. $y[n] = g[n]x[n]$, donde $g[n] = A sin(\omega n T)$ siendo $A$ constante, $\omega = 2 \pi f$ y $T$ el período de muestreo
2. $y[n] = \sum^{n+n_0}_{k=n-n_0} x[k]$
3. $y[n] = x[n] + 2$
4. $y[n] = nx[n]$

### 1
$$
\begin{align*}
y[n] &= g[n]x[n]\\
y[n] &= A sin(2 \pi f n T) x[n]
\end{align*}
$$

- Es causal porque solo depende del valor actual de entrada.
- Es lineal ya que cumple con la propiedad de superposición
$$
\begin{align*}
    x_1[n] \rightarrow y_1[n] \land x_2[n] \rightarrow y_2[n] \\
    \Rightarrow \alpha y_1[n] + \beta y_2[n] &\stackrel{?}{=} A \sin(2 \pi nfT) (\alpha x_1[n] + \beta x_2[n]) \\
    \alpha (A \sin(2 \pi nfT) x_1[n]) + \beta (A \sin(2 \pi nfT) x_2[n]) &\stackrel{?}{=} A \sin(2 \pi nfT) (\alpha x_1[n] + \beta x_2[n]) \\
    A \sin(2 \pi nfT)(\alpha x_1[n]) + \beta x_2[n]) &= A \sin(2 \pi nfT) (\alpha x_1[n] + \beta x_2[n])
\end{align*}
$$
- No es invariante en el tiempo, ya que un desplazamiento temporal de la entrada no produce el mismo desplazamiento de la salida
$$
\begin{align*}
    x[n-n_0] \rightarrow Asin(2 \pi nfT)x[n-n_0] &\neq y[n-n_0]\\
    Asin(2 \pi nfT)x[n-n_0] &\neq Asin(2 \pi (n-n_0)fT)x[n-n_0]\\
\end{align*}
$$
- No tiene memoria ya que la salida no depende de valores anteriores de entrada.

### 2
$$
\begin{align*}
y[n] &= \sum^{n+n_0}_{k=n-n_0} x[k]\\
n_0 = 2 \Rightarrow y[n] &= x[n-2] + x[n-1] + x[n] + x[n+1] + x[n+2]
\end{align*}
$$

- No es causal para $n_0 \neq 0$ porque depende de valores de entrada posteriores.
- Es lineal ya que cumple con la propiedad de superposición
$$
\begin{align*}
    x_1[n] \rightarrow y_1[n] \land x_2[n] \rightarrow y_2[n]\\
    \Rightarrow \alpha x_1[n] + \beta x_2[n] \rightarrow \sum^{n+n_0}_{k=n-n_0}
        \alpha x_1[k] + \beta x_2[k] &= \alpha y_1[n] + \beta y_2[n]\\
    \sum^{n+n_0}_{k=n-n_0} \alpha x_1[k] + \sum^{n+n_0}_{k=n-n_0} \beta x_2[k]
        &= \alpha \sum^{n+n_0}_{k=n-n_0} x_1[k] + \beta \sum^{n+n_0}_{k=n-n_0} x_2[k]\\
    \alpha \sum^{n+n_0}_{k=n-n_0} x_1[k] + \beta \sum^{n+n_0}_{k=n-n_0} x_2[k]
        &= \alpha \sum^{n+n_0}_{k=n-n_0} x_1[k] + \beta \sum^{n+n_0}_{k=n-n_0} x_2[k]
\end{align*}
$$
- Es invariante en el tiempo, ya que un desplazamiento temporal de la entrada produce el mismo desplazamiento de la salida
$$
\begin{align*}
    x[n-n_1] \rightarrow \sum^{n+n_0}_{k=n-n_0} x[k-n_1] &= y[n-n_1]\\
    \sum^{n+n_0-n_1}_{k=n-n_0-n_1} x[k] &= \sum^{n-n_1+n_0}_{k=n-n_1-n_0} x[k]
\end{align*}
$$
- Tiene memoria para $n_0 > 0$.

### 3
$$
\begin{align*}
y[n] &= x[n] + 2
\end{align*}
$$

- Es causal/no-anticipativo.
- El sistema no es lineal porque una entrada nula no anula la salida
$$
\begin{align*}
    x[n] = 0 \rightarrow y[n] = 2
\end{align*}
$$
- Es invariante en el tiempo, ya que un desplazamiento temporal de la entrada produce el mismo desplazamiento de la salida
$$
\begin{align*}
    x[n-n_0] \rightarrow x[n-n_0] + 2 &= y[n-n_0]
\end{align*}
$$
- El sistema no tiene memoria.

### 4
$$
\begin{align*}
y[n] &= nx[n]
\end{align*}
$$

- El sistema es causal.
- Es lineal ya que cumple con la propiedad de superposición
$$
\begin{align*}
    x_1[n] \rightarrow y_1[n] \land x_2[n] \rightarrow y_2[n]\\
    \Rightarrow \alpha x_1[n] + \beta x_2[n] \rightarrow n(\alpha x_1[n] + \beta x_2[n])
        &= \alpha y_1[n] + \beta y_2[n]\\
    n\alpha x_1[n] + n\beta x_2[n] &= \alpha nx_1[n] + \beta nx_2[n]
\end{align*}
$$
- No es invariante en el tiempo, ya que un desplazamiento temporal de la entrada no producen el mismo desplazamiento en la salida
$$
\begin{align*}
    x[n-n_0] \rightarrow nx[n-n_0] &\neq y[n-n_0]\\
    nx[n-n_0] &\neq (n-n_0)x[n-n_0]\\
    nx[n-n_0] &\neq nx[n-n_0] - n_0x[n-n_0]
\end{align*}
$$
- El sistema tampoco tiene memoria.

```python
import numpy as np
import matplotlib.pyplot as plt

'''
Impulso de magnitud 1 en t=x
'''
def impulso(tini,tfin,x,fm):
    # error si x esta fuera del rango
    if(x < tini or x >= tfin):
        raise Exception(f"x fuera de rango [tini,tfin) \nx={x}\ntini={tini}\ntfin={tfin}")

    t = np.linspace(tini,tfin,int(fm*(tfin-tini)),endpoint=False)
    
    # buscamos el t mas cercano (o igual) a x
    i = np.abs(t - x).argmin()
    x_t = t[i]

    return t, np.equal(t,[x_t]*len(t))

# Parametros
tini = 0
tfin = 2
fm = 25
T = 1.0/fm
# entrada (impulso unitario)
t,x1 = impulso(tini,tfin,1,fm)
x2 = np.sin(2*np.pi*t)

# salidas
# 1 f=1 A=1
y11 = np.multiply(np.sin(2*np.pi*t*T), x1)
y21 = np.multiply(np.sin(2*np.pi*t*T), x2)
# 2 - n0 = 10 El valor y2[k] es la sumatoria de x[k-n0] hasta x[k+n0]
n0=10
y12 = np.array([np.sum(x1[max(0,k-n0):min(len(x1)-1,k+n0)]) for k in range(len(x1))])
y22 = np.array([np.sum(x2[max(0,k-n0):min(len(x2)-1,k+n0)]) for k in range(len(x2))])
# 3
y13 = x1 + [2]*len(x1)
y23 = x2 + [2]*len(x2)
# 4
y14 = np.multiply(t,x1)
y24 = np.multiply(t,x2)

fig,ax = plt.subplots(5,2, sharex=True)
fig.set_figwidth(20)
fig.set_figheight(10)
for i in range(5):
    for j in range(2):
        ax[i][j].grid()

ax[0][0].set_title('$x_1[n]$')
ax[0][0].stem(t,x1)
ax[1][0].set_title('$x_1[n] \\rightarrow y_{11}[n]$')
ax[1][0].stem(t,y11)
# grafica de g[n]
#ax[1][0].plot(t,np.sin(2*np.pi*t*T))
ax[2][0].set_title('$x_1[n] \\rightarrow y_{12}[n]$')
ax[2][0].stem(t,y12)
ax[3][0].set_title('$x_1[n] \\rightarrow y_{13}[n]$')
ax[3][0].stem(t,y13)
ax[4][0].set_title('$x_1[n] \\rightarrow y_{14}[n]$')
ax[4][0].stem(t,y14)

ax[0][1].set_title('$x_2[n]$')
ax[0][1].stem(t,x2)
ax[1][1].set_title('$x_2[n] \\rightarrow y_{21}[n]$')
ax[1][1].stem(t,y21)
# grafica de g[n]
#ax[1][1].plot(t,np.sin(2*np.pi*t*T))
ax[2][1].set_title('$x_2[n] \\rightarrow y_{22}[n]$')
ax[2][1].stem(t,y22)
ax[3][1].set_title('$x_2[n] \\rightarrow y_{23}[n]$')
ax[3][1].stem(t,y23)
ax[4][1].set_title('$x_2[n] \\rightarrow y_{24}[n]$')
ax[4][1].stem(t,y24)
plt.show()
```
